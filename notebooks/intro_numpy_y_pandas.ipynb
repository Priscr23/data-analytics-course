# DOT PRODUCT

```python
def dot_product(a,b):
    result = 0
    for i in range(len(a)):
        result = result + a[i]*b[i]
    return result
```

```python
a = np.random.randn(100)
b = np.random.randn(100)
print(dot_product(a,b)) #Regresa un escalar
```

```python
import time

start = time.time()

total = time.time()-start
print('Total time: ', total )
```
