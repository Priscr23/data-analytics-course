```python
from sklearn.datasets import make_regression


X,y= make_regression(n_features=1, noise=40)
```

```python
X.shape, y.shape
```

```python
import pandas as pd 

df=pd.DataFrame(X)
df.columns=['X']
df['y']=y
df
```

```python
df['X'].plot(kind='hist', color='blue')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
 
y_mean=np.array([y.mean()]*100)
y_mean
```

```python
import matplotlib.pyplot as plt #clase el tipo de modelo

model_1=LinearRegression() #quiero un objeto de tipo linearRegression
model_2=LinearRegression()
model_3=LinearRegression()

model_1.fit(X,y)
```

```python
y_pred=model_1.predict(X)
```

```python
#Selector
plt.figure(figsize=(10,4)) #tamaño de la gráfica

#Data
plt.scatter(X,y, label="Random data")

plt.plot(X,y_mean, label="Mean", c="green")

plt.plot(X,y_pred, label="Linear Regression", c="red")

#show 
plt.xlabel("Feature")
plt.ylabel("Target (y)")
plt.title("Scatter example plot")
plt.legend()

plt.savefig("../reports/figures/random1.svg")
```
